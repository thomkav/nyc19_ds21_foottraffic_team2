```python
import datetime
import requests
import io
import pandas as pd
```

```python
# Set starting date
# Make sure date exists here: http://web.mta.info/developers/turnstile.html
date_val = datetime.date(2018, 9, 1)

# Adjust name of file, and output location
output_file_loc = '/Users/JacKuo14/Documents/Metis/nyc19_ds21_foottraffic_team2/data/mta_data.csv'

# Set range for number of weeks of data needed -- I added an extra week to make sure we had all days for a year
for x in range(53):
    
    url_1 = 'http://web.mta.info/developers/data/nyct/turnstile/turnstile_'
    url_2 = '.txt'
    date_val_str = date_val.strftime("%y%m%d") # convert datetime format to string
    
    # create url from pieces above
    full_url = url_1 + date_val_str + url_2
    
    # Get data from url
    r = requests.get(full_url)
    r.raise_for_status()
    
    # Create dataframe
    r_content = r.content
    df_mta_data = pd.read_csv(io.StringIO(r_content.decode('utf-8')))
    
    # Keep header if it's the first one, otherwise remove header and append to csv directly
    if x == 0:
        df_mta_data.to_csv(output_file_loc, index = False, header = True)
            
    else:
        with open(output_file_loc, 'a') as f:
            df_mta_data.to_csv(output_file_loc, mode = 'a', header = False, index = False)

    # Add 7 days to the date         
    date_val = date_val + datetime.timedelta(days = 7)
    
    # Just to make sure it's running smoothly
    print("Done with: " + full_url)
    print(x)

print("yass done done done!!")    
```
