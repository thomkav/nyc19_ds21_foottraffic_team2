# Load and filter the data

```python
brooklyn_df = pd.read_csv("../data/foodtruck_df.csv.zip")

top_10_stations = \
    list(brooklyn_df.groupby('UNIQUE_STATION').DAILY_TOTAL.sum().sort_values(ascending=False)[:10].index)

# Extract the data for the top 10 stations
top_10_df = brooklyn_df[brooklyn_df.UNIQUE_STATION.isin(top_10_stations)]

top_10_by_station_and_day = top_10_df.groupby(['UNIQUE_STATION', 'DAY_OF_WEEK']).DAILY_TOTAL.median()
# Turn the series back into a DataFrame
top_10_by_station_and_day = top_10_by_station_and_day.reset_index()

top_10_by_station_and_day
```

# Create the data table

```python
traffic_by_station_and_day = \
    pd.pivot_table(top_10_by_station_and_day, values='DAILY_TOTAL', index='UNIQUE_STATION', columns='DAY_OF_WEEK')

# Re-order the columns
traffic_by_station_and_day = \
    traffic_by_station_and_day[['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']]

traffic_by_station_and_day
```
